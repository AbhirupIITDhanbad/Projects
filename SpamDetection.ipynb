```python
import re
porter=PorterStemmer()
```

```python
df=text
```

```python
df=pd.DataFrame(df)
```

```python
def stemmer(text):
  stemming=re.sub('[^a-zA-Z]',' ',text)
  stemming=stemming.lower()
  stemming=stemming.split()
  stemming=[porter.stem(word) for word in stemming if not word in stopwords.words('english')]
  stemming=' '.join(stemming)
  return stemming
```

```python
df['text']=df['text'].apply(stemmer)
```
